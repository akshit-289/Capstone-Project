```python
data=pd.read_csv("rfmupdated.csv")
sale=pd.read_csv("updated.csv")
sale.columns
```

```python
sale['_Date'] = pd.to_datetime(sale['_Date'])

# Group by 'Customer ID' and find the maximum date for each group
max_date_per_customer = sale.groupby('Customer ID')['_Date'].max()

# Calculate the difference in days between the current date and the maximum date for each customer
current_date = pd.to_datetime()
churn_period = 90  # Churn period in days
days_since_last_purchase = (current_date - max_date_per_customer).dt.days

# Check if the difference exceeds the churn period
churned_customers = days_since_last_purchase > churn_period
print(list(churned_customers))
```
